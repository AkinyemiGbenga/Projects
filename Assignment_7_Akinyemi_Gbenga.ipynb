````python
# Import and load data
salary = "salary.csv"
data = pd.read_csv(salary)
print(f"Shape of Data: {data.shape}\n")

data.head()
````

````python
# Make copy of data
df = data.copy()
df.head()
````

````python
# Use pandas to make dummies for the data 

pd.get_dummies(df)
````

````python
#Show dummies of Salary - - <=50k and >50k

pd.get_dummies(df, columns=["Salary"])
````

````python
# Create the dummies dataframe
df_dummies = pd.get_dummies(df)
print(df_dummies.shape)

df_dummies.head()
````

````python
# Check the correlation
corr_matrix = df_dummies.corr()
corr_matrix
````

````python
# Spilt data into training and testing sets



RANDOM_STATE, TEST_SIZE = 5, 0.2
TARGET = "Salary_ >50K"

X = df_dummies.drop(columns=[TARGET])
y = df_dummies[TARGET]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
)

X_train.shape, X_test.shape
````

````python
# Check X_train
X_train.head(2)
````

````python
# Building Regression Model
# Instantiate
linear_model = LinearRegression()

# Fit
linear_model.fit(X_train, y_train)

# Make predictions
y_pred = linear_model.predict(X_test)
````

````python

np.array(y_pred)
````

````python
# create a dataframe for the prediction - returns dataframe with the true, predicted and error values


def result_dataframe(*, y_true: np.ndarray, y_pred: np.ndarray) -> pd.DataFrame:
    """This returns a DataFrame containing the true,
    predicted and error values."""
    # Convert predictions to a dataframe
    result = pd.DataFrame()  # Empty DF

    result["Actual_salary"] = y_true
    result["Predicted_salary"] = y_pred
    result["Error_in_predictions"] = (
        result["Actual_salary"] - result["Predicted_salary"]
    )

    return result
````

````python
# Question 1
result = result_dataframe(y_true=y_test, y_pred=y_pred)
result.sample(n=10, random_state=RANDOM_STATE)
````

````python
# Question 2
# Evaluate the performance of your model using at least three (3) performance metrics


# Mean Squared Error (The lower, the better)
mse = mean_squared_error(y_true=y_test, y_pred=y_pred, squared=True)

# Root Mean Squared Error (The lower, the better)
# Remember to use squared=False
rmse = mean_squared_error(y_true=y_test, y_pred=y_pred, squared=False)

# Mean Absolute Error (The lower, the better)
mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)

# R Squared (The higher, the better. Max (best) value is 1)
R2 = r2_score(y_true=y_test, y_pred=y_pred)

result_str = (
    "==================== Evaluation Metrics ===================="
    f"\nMean Squared Error: {mse}"
    f"\nRoot Mean Squared Error: {rmse}"
    f"\nMean Absolute Error: {mae}"
    "\n==========================================================="
    f"\nR Squared (The higher the better!): {R2} "
)

print(result_str)
````
